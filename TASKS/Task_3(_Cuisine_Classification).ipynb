```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Load the dataset
df = pd.read_csv('/content/Dataset .csv')
```

```python
# Handle missing values in the 'Cuisines' column by dropping rows with missing values
df_cleaned = df.dropna(subset=['Cuisines'])

# Convert 'Cuisines' to a categorical variable using label encoding
le = LabelEncoder()
df_cleaned['Cuisines_Encoded'] = le.fit_transform(df_cleaned['Cuisines'])
```

```python
# Select relevant features: City, Price range, Aggregate rating, and Votes
features = df_cleaned[['City', 'Price range', 'Aggregate rating', 'Votes']]
target = df_cleaned['Cuisines_Encoded']

# Encode the 'City' feature using one-hot encoding
features_encoded = pd.get_dummies(features, columns=['City'], drop_first=True)
```

```python
# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)
```

```python
# Initialize the Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Train the model
rf_classifier.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = rf_classifier.predict(X_test)
```

```python
# Generate the classification report
classification_report_output = classification_report(y_test, y_pred)
print('Classification Report:')
print(classification_report_output)
```
